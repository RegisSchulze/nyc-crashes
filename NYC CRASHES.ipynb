cleaning up vehicle_type_code1
ex. 'Ambulance' appears in different forms('AMB','AMBUL', ....) which should be replaced by 'Ambulance'

replace the all different reprensentations of Ambulance by 'Ambulance'

note that a better way would be to use regex, all values contain at least the letters a, m and b

the same could be done for other vehicle types and for colomns 'vehicle_type_code1' ==> 'vehicle_type_code5'

```python
df['vehicle_type_code1']=df['vehicle_type_code1'].replace({'AMB' : 'Ambulance', 
                                                            'AMBUL': 'Ambulance',
                                                            'Ambul': 'Ambulance',
                                                            'AMBU': 'Ambulance',
                                                            'AMBULANCE': 'Ambulance',
                                                            'GEN  AMBUL': 'Ambulance',
                                                            'AMBULACE': 'Ambulance',
                                                            'Leased amb': 'Ambulance',
                                                            'ambu': 'Ambulance',
                                                            'abulance': 'Ambulance',
                                                            'AMBULENCE': 'Ambulance',
                                                            'Fdny ambul': 'Ambulance',
                                                            'NYC AMBULA': 'Ambulance',
                                                            'White ambu': 'Ambulance',
                                                            'Amb': 'Ambulance'})
```

we are left with have 349 unique values in column 'vehicle_type_code1', a reduction of 15
